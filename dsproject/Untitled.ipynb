# PREDICTION OF PLAYERS USING DECISION TREE CLASSIFIER

Decision tree is a supervised learning technique which we can use in both regression as well as classification but in this project this algorithm is used for classification. Decision tree usually use multiple algorithms to decide to split a node into two or more nodes. It helps to evaluate the options and are excellent to choose between different course of actions.

When we will apply classification, the parameters will be x='Match_id', 'Team1', 'Team2', 'Season’, ‘Toss_decision', 'Toss_winner’ and 'Winner' whereas y will vary. For e.g. if we want to predict player 1 of them, we will take y= “Players_1” and train it and predict accuracy and after that y will be “Players_2”. We will predict 11 players name individually and calculate accuracy. After predicting all player names, we will calculate overall accuracy of model by taking average of accuracies calculated from players individually.


```python
x= df[['Match_id','Team1','Team2', 'Season','Toss_decision',
       'Toss_winner','Winner']]

x
```

# Player 1

```python
y= df['Players_1']
y

```

```python
from IPython import display

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=10)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

clf = DecisionTreeClassifier(random_state=1).fit(x_train, y_train)
clf.predict_proba(x_test[:11])

a1=(clf.score(x_test, y_test))
print("Accuracy of player 1=", a1 )
n1=clf.predict(x_test[:1])
print("Name of Player 1=",n1)

display.Image("https://img1.hscicdn.com/image/upload/f_auto,t_gn_f_345/lsci/db/PICTURES/CMS/316400/316489.png")
```
